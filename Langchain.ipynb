```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.prompts import PromptTemplate
from langchain.tools import Tool
from langchain.agents import initialize_agent, AgentType
from langchain.memory import ConversationBufferMemory
from langchain.chains import LLMChain
```

```python
llm=ChatGoogleGenerativeAI(model="gemini-1.5-pro", temprature=0.5,max_tokens=1000, google_api_key="AIzaSyBCxhmb7KM6YZy1fmJUCoLKcEM6NAuZcJU" )
```

```python
prompt=PromptTemplate(
    input_variables=["question"],
    template="You are a helpful AI assistant. Answer the following question: {question}"
)
```

```python
def search_tool(query):
    return f"Searching online for: {query}..."
search=Tool(name="Search",func=search_tool, description="A simple search tool")
```

```python
agent= initialize_agent(
tools=[search],
llm=llm,
agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
verbose=True
)
```

```python
memory=ConversationBufferMemory(memory_key="chat_history")
chain=LLMChain(llm=llm,prompt=prompt,memory=memory)
```

```python
import time
response1=(chain.run("What is Langchain?"))
print(response1)
time.sleep(3)
response2=chain.run("How does it compare to Open ai?")
print(response2)
```
